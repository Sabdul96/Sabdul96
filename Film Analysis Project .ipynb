## Libraries used 

```python
import pandas 
import seaborn 
import numpy 
import matplotlib
import matplotlib.pyplot as plt
matplotlib.pyplot.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
data = pandas.read_csv('Desktop/dataset/movies.csv')
```

```python
# Look for missing data 

for col in data.columns: 
    pct_missing = numpy.mean(data[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
data = data.dropna()
```

```python
# Creating correct Year column 
data['yearcorrect'] = data['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
data.head()
```

```python
data['budget'] = data['budget'].astype(int)
```

```python
data.sort_values(by=['gross'], inplace= False, ascending = False)
```

```python
pandas.set_option('display.max_rows', None)
```

```python
# Drops any production duplicates 
data['company'].drop_duplicates().sort_values(ascending = False)
data.tail()
```

```python
data.head()
```

# Analysis 

```python
# Scatterplot with budget vs gross 

plt.scatter(x = data['budget'], y=data['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot the budget vs gross using seaborn 

seaborn.regplot(x='budget', y='gross', data= data, scatter_kws = {"color": "red"}, line_kws = {"color": "black"})
;
```

```python
# Correlation
```

```python
data.corr( method = 'pearson') #pearson, kendall, spearman 
```

```python
# Viz for Correlation Matrix 

correlation_matrix = data.corr( method = 'pearson')
seaborn.heatmap (correlation_matrix, annot = True)

plt.title ('Correlation Matrix')
plt.xlabel ('Movie Features')
plt.ylabel (' Movie Features')

plt.show
; 
```

```python
# Looks at production company 

data.head()
```

```python
data_numerized = data 

for col_name in data_numerized.columns: 
    if(data_numerized[col_name].dtype == 'object'): 
        data_numerized[col_name] = data_numerized[col_name].astype('category')
        data_numerized[col_name] = data_numerized[col_name].cat.codes

data_numerized.tail()
```

```python
correlation_matrix = data_numerized.corr(method = 'pearson')
seaborn.heatmap (correlation_matrix, annot = True)

plt.title ('Correlation Matrix')
plt.xlabel ('Movie Features')
plt.ylabel (' Movie Features')

plt.show
; 
```

```python
data_numerized.corr()
```

```python
# Unstacking 

correlation_mat = data_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.6]

high_corr
```
